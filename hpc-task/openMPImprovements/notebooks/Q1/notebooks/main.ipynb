# Load Datasets with Results

```python
!pip install -q pandas matplotlib

import pandas as pd
import matplotlib.pyplot as plt

simple_df = pd.read_csv('../data/icx_simple_output.csv')
reduction_df = pd.read_csv('../data/icx_reduction_output.csv')
reduction_optimised_df = pd.read_csv('../data/icx_reduction_optimised_output.csv')
```

# Define Versions for visual legend and labels

```python
simple_df["Version"] = "Simple"
reduction_df["Version"] = "Reduction"
reduction_optimised_df["Version"] = "Reduction Optimised"
```

```python
df = pd.concat([simple_df, reduction_df, reduction_optimised_df])

plt.figure(figsize=(10, 6))
for label, group in df.groupby('Version'):
    print(df.head())
    plt.plot(group["Threads"], group["FuncCalls"], marker='o', label=label)

plt.title('Function Calls vs Threads for Different Versions of the OpenMP Code')
plt.xlabel('Number of Threads')
plt.ylabel('Number of Function Calls')
plt.grid(True)
plt.legend()
plt.savefig('../data/output/func_calls_vs_threads.png')
plt.show()
```
